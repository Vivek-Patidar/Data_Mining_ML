```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
data = pd.read_csv('/content/Market_Basket_Optimisation (2).csv')
data.shape
```

```python
transaction = []
for i in range(0, data.shape[0]):
    for j in range(0, data.shape[1]):
        transaction.append(data.values[i,j])

transaction = np.array(transaction)
```

```python
print(transaction)
```

```python
data.head()
```

```python
df = pd.DataFrame(transaction, columns=["items"])

df["incident_count"] = 1

indexNames = df[df['items'] == "nan" ].index
df.drop(indexNames , inplace=True)

df_table = df.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
df_table.head(5)
```

```python
transaction = []
for i in range(data.shape[0]):
    transaction.append([str(data.values[i,j]) for j in range(data.shape[1])])
transaction = np.array(transaction)
te = TransactionEncoder()
te_array = te.fit(transaction).transform(transaction)
df = pd.DataFrame(te_array, columns=te.columns_)
df.head()
```

```python
frequent_itemsets = fpgrowth(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets=association_rules(frequent_itemsets, metric="lift", min_threshold=1)
frequent_itemsets
```

```python
frequent_itemsets.sort_values("confidence",ascending=False)
```
